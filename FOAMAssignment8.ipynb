# **Use Graphical and Numerical Methode to solve**

a. $x^3-5x^2-2x+24 = 0$

b. $x^4+2x^3-3x^2-4x+4 = 0$

c. $ln(5x+7) = 5$

# **Solving Problem a using Graphical Methode**

>

```python
import numpy as np
from matplotlib import pyplot as plt


x_lower_limit = -5
x_upper_limit = 5
x = np.linspace(x_lower_limit, x_upper_limit, num=100)

y = x**3-5*x**2-2*x+24

y_lower_limit = -10
y_upper_limit = 30



fig, ax = plt.subplots()
ax.plot(x, y, )

plt.axhline(color='black')
plt.axvline(color='black')

plt.xlabel('$x$')
plt.ylabel('$y$')

ax.spines['top'].set_color('none')  #
ax.spines['bottom'].set_position('zero')

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')

ax.set_xlim([x_lower_limit, x_upper_limit])
ax.set_ylim([y_lower_limit, y_upper_limit])

ax.set_xticks(range(x_lower_limit, x_upper_limit, 1))
ax.set_yticks(range(y_lower_limit, y_upper_limit, 1))

plt.grid()
plt.show()
```

# **Solving Problem a using numberical methode**
>

```python
from scipy import optimize

roots = []


def f(x):
    return x**3-5*x**2-2*x+24


def find_solution():
    # if the function doesn't converge it will return a runtime error and the program will stop
    # Hence we are using a try and except block

    for guess in range(-100, 100):
        try:
            x0 = guess
            solution = round(optimize.newton(f, x0))

            if solution not in roots:
                roots.append(solution)


        except RuntimeError:
            print("No solution found")


if __name__ == "__main__":
    find_solution()

    print(roots)
```

# **Solving problem c using numerical methode**

```python
from scipy import optimize
import numpy as np

roots = []


def f(x):
    return np.log(5 * x + 7) - 5


def find_solution():
    for guess in range(-100, 100):
        try:
            x0 = guess
            solution = optimize.newton(f, x0)
            solution = round(solution, 2)

            if solution not in roots:
                roots.append(solution)
        except:
            pass


if __name__ == "__main__":
    find_solution()
    print(roots)
```

# **Question 2**

$y=x^2+3$ <br> $x^2+y^2=9$
>

```python
from sympy.solvers import solve
from sympy import Symbol

x = Symbol('x')
y = Symbol('y')

f_x_y = y - x ** 2 - 3
g_x_y = 9 - x ** 2 - y ** 2

solution = solve((f_x_y, g_x_y), x, y)

print(solution)
```
